```python
def load_and_test_on_oos(model_end_year, year_in_question, agrressive=False,
                         clip=False, model=None, tail_risk_fatter=False):
    if model is None:
        from_file = CatBoostClassifier()
        from_file.load_model(f'trained_model_upto_{model_end_year}.cbm')
        model = from_file

    X, y = simulate_oos(year= year_in_question, regression_dataset=False)
    if agrressive:
        preds = []
    
        for _, row in X.iterrows():
            temp_df = pd.DataFrame(row).T
            seed_diff = row['Diff_seeds']
            if seed_diff <= -12:
                # t1 is 12+ better:
                pred = [0,1]
            elif seed_diff >= 12:
                pred = [1,0]
            else:
                pred = model.predict_proba(temp_df)[0]
            preds.append(pred)
        #     return np.array(preds), y
    else:
        preds = model.predict_proba(X)
        
    if clip:
        preds = np.clip(preds , 0.05 , 0.95) 

    if tail_risk_fatter:
        l = []
        for prediction in preds:
            if 0.45 < prediction[0] < 0.55:
                choice = np.random.randint(0,2)
                if choice == 0:
                    new_prediction = [.6,.4]
                elif choice == 1:
                    new_prediction = [.4,.6]
            else:
                new_prediction = prediction
            l.append(new_prediction)
        preds = l
        
    return log_loss(y_true = np.array(y), y_pred=np.array(preds))
```

```python
def predict_proba(score_diff_est, sigma):
    t1_wins = norm.cdf(- (score_diff_est - mu) / sigma)
    t2_wins = 1 - t1_wins
    
    return np.array(list(zip(t1_wins, t2_wins)))
```

```python
def load_and_test_on_oos_regression(model_end_year, year_in_question, agrressive=False,
                         clip=False, model=None, tail_risk_fatter=False, sigma=15):
    X, y = simulate_oos(year= year_in_question, regression_dataset=True)
    
    if agrressive:
        preds = []
    
        for _, row in X.iterrows():
            temp_df = pd.DataFrame(row).T
            seed_diff = row['Diff_seeds']
            if seed_diff <= -12:
                # t1 is 12+ better:
                pred = [0,1]
            elif seed_diff >= 12:
                pred = [1,0]
            else:
                pred = predict_proba(model.predict(temp_df), sigma=sigma)[0]
            preds.append(pred)
        #     return np.array(preds), y
    else:
        preds = predict_proba(model.predict(X), sigma=sigma)
        
    if clip:
        preds = np.clip(preds , 0.05 , 0.95) 

    if tail_risk_fatter:
        l = []
        for prediction in preds:
            if 0.45 < prediction[0] < 0.55:
                choice = np.random.randint(0,2)
                if choice == 0:
                    new_prediction = [.6,.4]
                elif choice == 1:
                    new_prediction = [.4,.6]
            else:
                new_prediction = prediction
            l.append(new_prediction)
        preds = l
    
    y_true = []
    for el in y:
        if el < 0:
            y_t = [1,0]
        else:
            y_t = [0,1]
        y_true.append(y_t)
    y_true = np.array(y_true)
    
    return log_loss(y_true = y_true, y_pred=np.array(preds))
```
