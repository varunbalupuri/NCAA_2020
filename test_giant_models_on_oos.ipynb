```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from catboost import CatBoostClassifier, Pool, cv
```

```python
for end_year in [2015,2016,2017,2018]:
    X_train, y_train, X_test, y_test = train_giant_model(end_year=end_year)
    categorical_idxs = [0, 1]
    model = CatBoostClassifier(custom_loss=['Logloss'], logging_level='Verbose')
    model.fit(X_train, y_train, cat_features=categorical_idxs, eval_set=(X_test, y_test), plot=False)

    feature_importances = pd.Series(dict(zip(X_train.columns,
                                             model.get_feature_importance()))).sort_values(ascending=False)
    scores = model.get_best_score()
    model.save_model(f'trained_model_upto_{end_year}.cbm')
    print(f'saved model trained up to : {end_year}')
```

```python
from sklearn.metrics import log_loss
```

```python
from benchmark_model import simulate_oos
```

```python
def load_and_test_on_oos(model_end_year, year_in_question):
    from_file = CatBoostClassifier()
    from_file.load_model(f'trained_model_upto_{model_end_year}.cbm')
    
    X, y = simulate_oos(year= year_in_question)
    predictions = from_file.predict_proba(X)
    predictions = np.clip(predictions , 0.05 , 0.95)
    return log_loss(y_true = np.array(y), y_pred=np.array(predictions))
```

```python
for model_end_year in [2015,2016,2017,2018]:
    for year_in_question in [i for i in range(model_end_year+1, 2020)]:
        print(f'model_end_year={model_end_year}, year_in_question={year_in_question}, log_loss={load_and_test_on_oos(model_end_year, year_in_question)}')
```
